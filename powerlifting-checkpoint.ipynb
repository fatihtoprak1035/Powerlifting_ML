**WeightClassKg sutunundaki unique degerlere bakiyorum. Bu degerlere gore en yakin classa yuvarlama islemi yapacagim. Kisacasi BodyweightKg'da ki degeri kullanarak WeightClassKg'i en yakin ust deger olan class olarak belirleyecegim.**

```python
unique_values = train['WeightClassKg'].unique()
print(f'We Have {len(unique_values)} Unique values. Values in the WeightClassKg Column: {unique_values}')
print('__'*30)
```

```python
def assign_weight_class(bodyweight):
    # Eger bodyweight NaN ise NaN dondurur.
    if pd.isna(bodyweight):
        return np.nan

    weight_classes = [
        36, 40, 43, 44, 46, 47, 48, 49, 50, 52, 52.5, 54.4, 55, 55.5, 56, 56.7, 57, 58.5, 58.9, 59, 
        60, 63, 63.5, 66, 67, 67.5, 68, 68.5, 68.9, 69.5, 70, 70.3, 70.5, 71.5, 72, 72.5, 73.5, 74, 
        74.5, 75, 75.5, 76.5, 77.5, 78.5, 78.9, 79.3, 79.5, 80, 80.5, 81.6, 82, 82.5, 83, 83.9, 84, 
        84, 90, 90.7, 93, 100, 102, 103, 105, 110, 113.4, 117.5, 117.9, 120, 125, 127, 136, 140, 
        145, 155, 110, 90, 75, 82.5, 52, 125, 140, 145, 72, 93, 60, 80, 67.5, 105, 
        63, 155, 78.9, 117.9, 113.4
    ]
    
    # Duplicateleri kaldirir ve her ogeyi float tipine cevirip sirali bir kumeye donusturur.
    weight_classes = sorted(set(float(str(x).replace('+', '')) for x in weight_classes if x != '+'))
    
    # Vucut agirliginin altinda ya da esit olan ilk kilo sinifini bulup donderir ( ͡~ ͜ʖ ͡°).
    for threshold in weight_classes:
        if bodyweight <= threshold:
            return str(threshold)
    
    # Eger vucut agirligi tum siniflardan buyukse en yuksek sinifi "+" isaretiyle dondurur.
    return f"{weight_classes[-1]}+"

def fill_weight_class(train):
    # 'BodyweightKg' sutununu kullanarak, 'WeightClassKg' sutununu assign_weight_class fonksiyonu ile doldurur.
    train['WeightClassKg'] = train['BodyweightKg'].apply(assign_weight_class)
    return train
```

**AgeClass sütunundaki boş değerleri doldurmak için öncelikle AgeClass'ın unique değerlerine bakacağım. Bu değerlere göre, boş olanları Age sütununun değerlerine dayalı olarak otomatik olarak sınıflandıracağım.**

```python
unique_values = train['AgeClass'].unique()
print(f'We Have {len(unique_values)} Unique values. Values in the AgeClass Column: {unique_values}')
print('__'*30)
```

```python
def get_age_class(age):
    if pd.isna(age):
        return np.nan
    
    age_ranges = [
        (5, 12, '5-12'),
        (13, 15, '13-15'),
        (16, 17, '16-17'),
        (18, 19, '18-19'),
        (20, 23, '20-23'),
        (24, 34, '24-34'),
        (35, 39, '35-39'),
        (40, 44, '40-44'),
        (45, 49, '45-49'),
        (50, 54, '50-54'),
        (55, 59, '55-59'),
        (60, 64, '60-64'),
        (65, 69, '65-69'),
        (70, 74, '70-74'),
        (75, 79, '75-79'),
        (80, 999, '80-999')
    ]
    
    # Yas araliklari icinde belirtilen yas varsa o yasa karsilik gelen sinifi dondurur
    for start, end, age_class in age_ranges:
        if start <= age <= end:
            return age_class
    return np.nan

def fill_age_class(train):
    # 'AgeClass' sutununda bos olanlar icin, 'Age' sutunundaki degerleri kullanarak 'AgeClass' degerini doldurur
    train.loc[train['AgeClass'].isna(), 'AgeClass'] = train.loc[train['AgeClass'].isna(), 'Age'].apply(get_age_class)
    return train
```

## XGBoost Hyperparameter bulmaca

```python
import optuna
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def objective(trial):
    # Define hyperparameters
    n_estimators = trial.suggest_int('n_estimators', 50, 200)  # Number of trees
    max_depth = trial.suggest_int('max_depth', 3, 10)  # Maximum depth of the trees
    learning_rate = trial.suggest_float('learning_rate', 1e-3, 1, log=True)  # Learning rate
    subsample = trial.suggest_float('subsample', 0.5, 1.0)  # Subsample ratio

    # Create and fit the XGBoost regressor
    regressor_xgb = XGBRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        subsample=subsample,
        random_state=42
    )
    
    regressor_xgb.fit(X_train, y_train)

    # Make predictions on the validation/test data
    y_pred = regressor_xgb.predict(X_test)

    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # We aim to minimize RMSE, so we return it as the optimization target
    return rmse

# Start the Optuna study
study = optuna.create_study(direction='minimize')  # Minimize RMSE
study.optimize(objective, n_trials=50)  # Run 50 trials

# Print the best parameters and RMSE
print('Best parameters:', study.best_params)
print('Best RMSE:', study.best_value)

# Evaluate with the best model
best_params = study.best_params
best_regressor_xgb = XGBRegressor(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    learning_rate=best_params['learning_rate'],
    subsample=best_params['subsample'],
    random_state=42
)

best_regressor_xgb.fit(X_train, y_train)
best_y_pred = best_regressor_xgb.predict(X_test)

# Calculate final metrics
final_rmse = np.sqrt(mean_squared_error(y_test, best_y_pred))
final_r2 = r2_score(y_test, best_y_pred)

# Print final metrics
print("Final RMSE:", final_rmse)
print("Final R² Score:", final_r2)
```
